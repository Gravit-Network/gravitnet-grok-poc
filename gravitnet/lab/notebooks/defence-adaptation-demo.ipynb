```python
# Setup
from runtime import GravitEngine, DefenseAdapter

engine = GravitEngine(num_nodes=50)
engine.defense_adapter = DefenseAdapter()  # enable adaptation

history_static = {"honest": [], "adv": [], "q_rate": []}
history_adapt = {"honest": [], "adv": [], "q_rate": [], "theta_q": [], "scar_impact": []}
```

```python

# Run static (no adaptation)
for t in range(60):
    engine.step()  # run without adaptation
    history_static["honest"].append(engine.history["honest_avg_tau"][-1])
    history_static["adv"].append(engine.history["adversary_avg_tau"][-1])
    history_static["q_rate"].append(engine.history["quarantine_rate"][-1])
```

```python
# Run adaptive
engine = GravitEngine(num_nodes=50)
engine.defense_adapter = DefenseAdapter()

for t in range(60):
    engine.step()
    history_adapt["honest"].append(engine.history["honest_avg_tau"][-1])
    history_adapt["adv"].append(engine.history["adversary_avg_tau"][-1])
    history_adapt["q_rate"].append(engine.history["quarantine_rate"][-1])
    history_adapt["theta_q"].append(engine.defense_adapter.theta_q)
    history_adapt["scar_impact"].append(engine.defense_adapter.scar_impact)
```

```python
# Plot
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(history_static["honest"], label="Honest (static)")
plt.plot(history_static["adv"], label="Attacker (static)")
plt.title("Static Defense")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(history_adapt["honest"], label="Honest (adaptive)")
plt.plot(history_adapt["adv"], label="Attacker (adaptive)")
plt.title("Adaptive Defense")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(history_adapt["theta_q"], label="θ_q")
plt.plot(history_adapt["scar_impact"], label="Scar Impact")
plt.title("Defense Parameters Evolution")
plt.legend()

plt.tight_layout()
plt.show()
```
