```python
# Cell 1 — Setup

from runtime import GravitEngine
import matplotlib.pyplot as plt

engine = GravitEngine(num_nodes=40)
```

```python
# Cell 2 — Run simulation

steps = 30
honest_history = []
sybil_history = []

for t in range(steps):
    engine.step()

    honest = []
    sybil = []

    for i, node in engine.network.nodes.items():
        tau = node.compute_trust()
        if i in engine.network.sybil_nodes:
            sybil.append(tau)
        else:
            honest.append(tau)

    if honest:
        honest_history.append(sum(honest)/len(honest))
    if sybil:
        sybil_history.append(sum(sybil)/len(sybil))
```

```python
# Cell 3 — Plot

plt.plot(honest_history, label="Honest")
plt.plot(sybil_history, label="Sybil")

plt.title("Sybil Attack Simulation")
plt.xlabel("Time step")
plt.ylabel("Average Trust")

plt.legend()
plt.show()
```

```python
# Run simulation with signals
engine = GravitEngine(num_nodes=40)

honest_history = []
sybil_history = []

for t in range(30):
    engine.step()

    honest = []
    sybil = []

    for i, node in engine.network.nodes.items():
        tau = node.compute_trust()
        if i in engine.network.sybil_nodes:
            sybil.append(tau)
        else:
            honest.append(tau)

    honest_history.append(sum(honest)/len(honest))
    sybil_history.append(sum(sybil)/len(sybil))
```

```python
# Plot
import matplotlib.pyplot as plt

plt.plot(honest_history, label="Honest")
plt.plot(sybil_history, label="Sybil")

plt.title("Signal Injection under Sybil")
plt.xlabel("Time")
plt.ylabel("Trust")

plt.legend()
plt.show()
```

```python
collusion_history = []

for t in range(30):
    engine.step()

    collusion = []
    for i, node in engine.network.nodes.items():
        if i in engine.network.collusion_nodes:
            collusion.append(node.compute_trust())

    collusion_history.append(sum(collusion)/len(collusion))

plt.plot(collusion_history, label="Collusion")
```
